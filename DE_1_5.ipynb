#Библиотеки, загрузка датасета

```python
import pandas as pd

df = pd.read_csv("/content/ds_salaries.csv")
df
```

#Задания

**Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)**

---

```python
df.groupby("job_title")["salary_in_usd"].count()
```

**Какая средняя и медианная зарплата по группам вакансий?**

---

```python
df.groupby("job_title")["salary_in_usd"].agg(['mean', 'median']).round({'mean':2, 'median':2})
```

**Какая средняя и медианная зарплата по каждому региону?**

---

```python
df.groupby("employee_residence")["salary_in_usd"].agg(['mean', 'median']).round({'mean':2, 'median':2})
```

**Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?**

---

```python
df.groupby("job_title")["salary_in_usd"].agg(['mean']).sort_values(by=["mean"], ascending=False).head(1)
```

**Какое процентное соотношение каждого региона по вакансиям от всех вакансий?**

---



```python
df_2 = pd.DataFrame(df.groupby("employee_residence")["salary_in_usd"].count(), columns=["salary_in_usd"])
df_2["residence_percentage"] = df_2.apply(lambda x: round(x * 100 / df.count()[0], 2), axis=1)

df_2.sort_values(by=["salary_in_usd"], ascending=False)["residence_percentage"]
```

**Какая корреляция уровня опыта от зарплаты?**

---


```python
experience_levels = {
    'EN': 1,
    'MI': 2,
    'SE': 3,
    'EX': 4,
}

df_2 = pd.DataFrame(df, columns=["salary_in_usd", "experience_level"])
df_2["experience_level_num"] = df_2.apply(lambda x: experience_levels[x["experience_level"]], axis=1)
df_3 = pd.DataFrame(df_2, columns=["experience_level_num", "salary_in_usd"])

df_3.corr().loc["salary_in_usd", "experience_level_num"]
```
